```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft
```

#**1**
Transformada Rápida de Fourier para señales discretas

```python
def dftmatrix(N, Nfft):
    #construct DFT matrix
    k= np.arange(Nfft)
    if N is None: N = Nfft
    n = np.arange(N)
    U = np.matrix(np.exp(1j* 2*np.pi/Nfft *k*n[:,None]))
    return U/np.sqrt(Nfft)
```


#**2.1.**
Teniendo un sistema con fs = 2kHz, cree un vector tiempo de 0.8 segundos y genere las siguientes señales de acuerdo con el último número de su cédula (C).

```python
C = 2 #último dígito de la cedula
fs = 2000
t1 = np.arange(0, 0.8, 1/fs)#vector tiempo

#Párametros adicionales
f1 = (1 + C)*15
f2 = 2*C + 20
f3 = (2 + C)*50 - 90
f4 = 4*C + 80

#Señales

x1 = np.cos(2*np.pi*f1*t1)
x2 = np.cos(2*np.pi*f2*t1)
x3 = np.cos(2*np.pi*f3*t1)
x4 = np.cos(2*np.pi*f4*t1)

print (f'F1 = {f1}')
print (f'F2 = {f2}')
print (f'F3 = {f3}')
print (f'F4 = {f4}')
```


#**2.1**

#**2.3**
Calcule la Transformada de Fourier usando U y Nfft = 4096 para cada señal.

```python
nfft = 4096
N = len(x1)
freq=np.hstack((np.arange(0, nfft/2-1), np.arange(-nfft/2, 1)))*fs/nfft # vector de frecuencias ...para poder graficar la fft en Hz

#Para x1
x1.shape=(len(x1),1) # Convierto en vector columna para poder multiplicar con la matriz
U1=dftmatrix(len(x1), nfft) # calculo la matriz de transformacion
A1=U1.H*x1[:] # calculo la fft

#Para x2
x2.shape=(len(x2),1) # Convierto en vector columna para poder multiplicar con la matriz
U2=dftmatrix(len(x2), nfft) # calculo la matriz de transformacion
A2=U2.H*x2[:] # calculo la fft

#Para x3
x3.shape=(len(x3),1) # Convierto en vector columna para poder multiplicar con la matriz
U3=dftmatrix(len(x3), nfft) # calculo la matriz de transformacion
A3=U3.H*x3[:] # calculo la fft

#Para x4
x4.shape=(len(x4),1) # Convierto en vector columna para poder multiplicar con la matriz
U4=dftmatrix(len(x4), nfft) # calculo la matriz de transformacion
A4=U4.H*x4[:] # calculo la fft

Ux = dftmatrix(N,nfft)
```

* Para calcular la Transformada Rápida de Fourier (FFT), es necesario generar una matriz de transformación en la que se asegure que el número de muestras (N) coincida con el número de puntos de la FFT (Nfft). Sin embargo, en este escenario, la longitud del segmento de la señal no es uniforme en todos los casos, ya que se está utilizando una Nfft constante de 4096.
